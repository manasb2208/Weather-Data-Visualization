```python
import matplotlib.pyplot as plt
import pandas as pd
```

Taking two databases for same month from two different location and focus on 
try to find the data variation

```python
df=pd.read_csv("/content/kolkata_january_2020_temp_database.csv")
```

```python
df1=pd.read_csv("/content/mumbai_january_2020_temp_database.csv")
```

```python
df.head()
```

```python
df1.head()
```

```python
plt.plot(df['Date'],df['Highest_Temp'])


plt.title("Date vs High Temperature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High Temp. in degree celcius",fontsize=15)
```

```python
plt.plot(df['Date'],df['Lowest_Temp'])


plt.title("Date vs Low Temperature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Low Temp. in degree celcius",fontsize=15)
```

So, We see that maximum  Highest Temperature in kolkata in January2020 was 30 degree celcius and maximum lowest temperature was 19 degree celcius

We also see that in the mid of month lowest temperature arises in the corresponding temperature range

#Plotting in a one space

```python
plt.subplot(2,1,1)
plt.plot(df['Date'],df['Highest_Temp'])


plt.title("Date vs High Temperature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High Temp. in degree celcius of kolkata",fontsize=15)

plt.show()



plt.subplot(2,1,2)
plt.plot(df['Date'],df['Lowest_Temp'])


plt.title("Date vs Low Temperature",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Low Temp. in degree celcius of kolkata",fontsize=15)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(df['Date'],df['Highest_Temp'])


plt.title("Kolkata",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High Temp. in degree celcius of kolkata",fontsize=15)

plt.show()



plt.subplot(2,1,2)
plt.plot(df1['Date'],df1['Highest_Temp'])


plt.title("Mumbai",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("High Temp. in degree celcius of Mumbai",fontsize=15)

plt.show()
```

Wee see here maxixmum highest temperature over the january month in mumbai was 35 degree celcius and that of kolkata is around 30 degree celcius, so we see that Mumbai is respectively warmer place than Kolkata

```python
plt.subplot(2,1,1)
plt.plot(df['Date'],df['Lowest_Temp'])


plt.title("Kolkata",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Low Temp. in degree celcius of kolkata",fontsize=15)

plt.show()



plt.subplot(2,1,2)
plt.plot(df1['Date'],df1['Lowest_Temp'])


plt.title("Mumbai",fontsize=20)
plt.xlabel("Date",fontsize=15)
plt.ylabel("Low Temp. in degree celcius of Mumbai",fontsize=15)

plt.show()
```

We see here maxixmum lowest temperature over the january month in mumbai was 28 degree celcius and that of kolkata is around 17.5 degree celcius

Lowest Temperature per day over the month in mumbai doesn't follow any pattern respect to that of kolkata. In case of Kolkata it is seen that at the end of month temperature is slowly increasing
